```python
import numpy as np
from scipy.optimize import minimize

import matplotlib.pyplot as plt
import matplotlib.cm as cm
from mpl_toolkits.mplot3d import Axes3D

```

```python
def f(x, y):
    """Himmelblau’s function"""
    
    return (x**2 + y - 11)**2 + (x + y**2 - 7)**2
```

```python
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)

xx, yy = np.meshgrid(x, y)

z = f(xx, yy)
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(1, 1, 1, projection='3d')
surf = ax.plot_surface(xx, yy, z, cmap=cm.viridis, rstride=2, cstride=2)

# Add labels and a color bar
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis (function output)')
fig.colorbar(surf, shrink=0.5, aspect=10)

plt.title('3D Surface Plot of a Function with Two Arguments')
plt.show()
```

```python
fig, ax = plt.subplots()

im = ax.imshow(z, cmap=cm.RdBu, vmin=abs(z).min(),
     vmax=abs(z).max(), extent=[0, 1, 0, 1])
im.set_interpolation('bilinear')

cb = fig.colorbar(im, ax=ax)
```
