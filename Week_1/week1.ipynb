```python
data = np.array([1.77, -0.23, 2.76, 3.80, 3.47, 56.75, -1.34, 4.24, -2.44, 3.29, 3.71, -2.40, 4.53, -0.07, -1.05, -13.87, -2.53, -1.75, 0.27, 43.21])
```

```python
prep_data = np.zeros([len(data), 2])
prep_data[:,0] = data

plt.plot((-20, 60), (0, 0), "-")
plt.plot(prep_data[:, 0], prep_data[:, 1], '.', markersize="20")
plt.show()
```

```python
def cauchy_pdf(x_data, theta, gamma=1.0):
    """
    theta is the mode
    gamma is the scale param
    """

    x0 = theta
    x = x_data

    return 1.0/(np.pi*gamma*(1 + ((x - x0)/(gamma))**2))
```

```python
thetas = np.linspace(-5, 5, 1000)

Ls = np.zeros(len(thetas))

for i, theta in enumerate(thetas):
    Ls[i] = (np.prod(cauchy_pdf(data, theta=theta)))

plt.plot(thetas, np.log(Ls), "-")
```

# Now using numpy broadcasting to create matrix of x_data values and theta values, can then skip the for-loop. ALSO calculating the log of the matrix before summing (collapsing) in the x_data dimension. This is better for numerical stability compared to first multiplying the matrix elements before and then taking the log.

```python
plt.plot(thetas, np.sum(np.log(cauchy_pdf(data.reshape(-1,1), thetas.reshape(1,-1))), axis=0))
```

```python
def newtons_method(x_start, dg, ddg):
    x_current = x_start 
    x_new = x_current - (dg/ddg)
    return x_new
```
